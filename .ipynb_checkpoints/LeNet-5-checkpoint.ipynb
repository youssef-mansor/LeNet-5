## Convolutional Module 2d 

```python
from torch import nn
```

```python
class Conv2d(d2l.HyperParameters, nn.Module):
    
    def __init__(self, output_channels, padding, kernal_size = 5):
        super().__init__()
        self.save_hyperparameters()
        self.kernel = nn.Parameter(data = torch.randn((5, 5)), requires_grad=True)
    
    def forward(self, X): #shape (batch, channels, height, width)
        h = X.shape[0]
        w = X.shape[1]
        padded_X = torch.zeros((h + 2*self.padding , w + 2*self.padding))
        for i in range(h):
            for j in range(w):
                padded_X[i+self.padding][j+self.padding] = X[i][j]
        return padded_X
```

```python
conv2d = Conv2d(1, 2, 5)
X = torch.randn((2, 2))
```
